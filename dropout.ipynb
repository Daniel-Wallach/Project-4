```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("data.csv",sep=";")
df
```

```python
# print the shape of the DataFrame, the number of rows and columns

print(f'The data frame has {df.shape[0]} rows and {df.shape[1]} columns.')
```

```python
# show the general information of the data

df.info()
```

```python
# check if there are null values in the data

df.isnull().sum()
```

```python
# general statistics of the data

df.describe()
```

```python
import seaborn as sns
import matplotlib.gridspec as gridspec

# plot the distribution of each variable

fig = plt.figure(figsize=(20, 20))
# give more space between the plots
fig.subplots_adjust(hspace=0.5, wspace=0.4)
gs = gridspec.GridSpec(10, 4)

# foor loop to add each plot (remove the last column because it is the target variable)
for i, col in enumerate(df.columns[:-1]):
    ax = fig.add_subplot(gs[i])
    sns.histplot(df[col], ax=ax)
    ax.set_xlabel('')
    ax.set_title('feature: ' + col)

plt.show()
```

```python
# correlation analysis between variables
# remove the last column because it is the target variable

corr = df.iloc[:, :-1].corr()
corr
```

```python
# Plotting a correlation matrix 
fig = plt.figure(figsize=(10, 8))

sns.heatmap(df.corr(),annot=False, cmap='RdBu', center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show
```

```python
import textwrap
```

```python
# plot the relationship betwee each variables and the target variable (which is the last column). Using boxplot

fig = plt.figure(figsize=(20, 20))
# give more space between the plots
fig.subplots_adjust(hspace=0.6, wspace=0.5)
gs = gridspec.GridSpec(10, 4)

# foor loop to add each plot (remove the last column because it is the target variable)
for i, col in enumerate(df.columns[:-1]):
    ax = fig.add_subplot(gs[i])
    sns.boxplot(x=df['Target'], y=df[col], ax=ax)
    ax.set_xlabel('')
    # wrap the y labels
    ax.set_ylabel(textwrap.fill(col, 15))
    #ax.set_title('feature: ' + col)

plt.show()
```
